```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import numpy as np
```

```python
# Generating some random data
np.random.seed(42)
X=np.random.rand(100,10)
y=np.random.rand(100,1)

# Define a simple model
def create_model(loss):
    model=Sequential([
        Dense(64,input_dim=10,activation='relu'),
        Dense(64,activation='relu'),
        Dense(1,activation='linear'),
        
    ])
    model.compile(optimizer=Adam(learning_rate=0.01),loss=loss,metrics=['mae'])
    return model

# List of loss fuctions to demonstrate
loss_functions={
    'Mean Squared Error':'mean_squared_error',
    'Mean Absolute Error':'mean_absolute_error',
    'Mean Squared Logarithmic Error':'mean_squared_logarithmic_error',
}

    
    
```

```python
# Traning the model with diffrent loss fuction
history={}
for name,loss in loss_functions.items():
    print(f"Traning with {name} loss function")
    model=create_model(loss)
    history[name]=model.fit(X,y,epochs=100,batch_size=10,verbose=0)
    print(f"Finished training with {name} loss function \n")
```

```python
# Plotting the results
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
for name in loss_functions.keys():
    plt.plot(history[name].history["loss"],label=f"{name} Loss")
plt.title("Loss per Epoch for Diffrent Loss functions")
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
