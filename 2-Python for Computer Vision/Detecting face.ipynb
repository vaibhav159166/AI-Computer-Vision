### Face Detection with Haar Cascades Note: This is face detection NOT face recognition. We are only detecting if a face is in an image, not who the face actually is. That requires deep learning which we'll go over later!

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

https://en.wikipedia.org/wiki/Denis_Mukwege<br>

```python
nadia=cv2.imread("Nadia_Murad.jpg",0)
denis=cv2.imread("Denis_Mukwege.jpg",0)
solvay=cv2.imread("solvay_conference.jpg",0)
```

```python
plt.imshow(solvay,cmap='gray')
```

```python
face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
def detect_face(img):
    face_img=img.copy()
    face_rects=face_cascade.detectMultiScale(face_img)
    
    for(x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
        return face_img
```

```python
result=detect_face(denis)
```

```python
plt.imshow(result,cmap='gray')
```

```python
result=detect_face(nadia)
plt.imshow(result,cmap='gray')
```

```python
result1=detect_face(solvay)
plt.imshow(result1,cmap='gray')
```
